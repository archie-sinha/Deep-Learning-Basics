```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

##1. LINEAR FUNCTION

```python
# EQUATION
a = random.randint(1,10)
c = random.randint(1,10)
x= random.randint(1,10)
lin_fn= (a*x) + c
print("Linear fn: ", lin_fn)
# DERIVATIVE
der_lin_fn= a*x
print("Derivative of Linear fn: ", der_lin_fn)
```

```python
#Graph of Linear Function
x =np.arange(-5,5,0.1)
lin_fn= (a*x) + c
der_lin_fn= a*x

plt.title("Linear Graph")
plt.plot(x,lin_fn)
plt.plot(x,der_lin_fn)
plt.show()
```

##2. SIGMOID FUNCTION

```python
# EQUATION
x =np.arange(-5,5,0.1)
sig_fn = 1/(1+np.exp(-x))

# DERIVATIVE
der_sig_fn=sig_fn*(1-sig_fn)

plt.title("Sigmoid graph")
plt.plot(x,sig_fn)
plt.plot(x,der_sig_fn)
plt.show()
```

##3. TANH FUNCTION

```python
# EQUATION
x =np.arange(-5,5,0.1)
tanh_fn = (2/(1+np.exp(-2*x)))-1

# DERIVATIVE
der_tanh_fn= 1- (tanh_fn**2)

plt.title("Tanh graph")
plt.plot(x,tanh_fn)
plt.plot(x,der_tanh_fn)
plt.show()
```

##4. RELU FUNCTION

```python
# EQUATION
x = np.arange(-5, 5, 0.1)

relu_fn = []

for i in x:
    if i < 0:
        relu_fn.append(0)
    else:
        relu_fn.append(i)

# DERIVATIVE
der_relu_fn = []

for i in x:
    if i < 0:
        der_relu_fn.append(0)
    else:
        der_relu_fn.append(1)

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.title("ReLU Function")
plt.plot(x, relu_fn)
plt.xlabel("x")
plt.ylabel("ReLU(x)")

plt.subplot(122)
plt.title("Derivative of ReLU")
plt.plot(x, der_relu_fn)
plt.xlabel("x")
plt.ylabel("ReLU'(x)")

plt.tight_layout()
plt.show()



```
