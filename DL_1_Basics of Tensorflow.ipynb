# Task 1: Perform basic operations of Tensorflow


```python
#version
import numpy as np
import tensorflow as tf
tf.__version__
```

```python
#constants
x = tf.constant([[1,2,3],[4,5,6]])
print(x)
print(x.shape)
print(x.dtype)
print(x.numpy())
```

```python
#variables
var = tf.Variable([0.0,0.0,0.0])
var.assign([1,2,3])
print(var)
var.assign_add([1,1,1])
print(var)
print(var.numpy())
```

```python
#concatenation on columns
A = tf.constant([[3,4,5],[1,2,3]])
B = tf.Variable([[1,1,1],[2,3,4]])
AB_conc_col = tf.concat([A,B],axis=1)
print(AB_conc_col.numpy())
```

```python
#concatenation on rows
A = tf.constant([[3,4,5],[1,2,3]])
B = tf.Variable([[1,1,1],[2,3,4]])
AB_conc_row = tf.concat([A,B],axis=0)
print(AB_conc_row.numpy())
```

```python
#reshape
t1 = tf.reshape(AB_conc_row,shape = [1,12])
print(t1.numpy())
```

```python
#operations
a = tf.constant(2.0)
b = tf.constant(3.5)
c = tf.constant(5.6)
add  = tf.add(tf.add(a,b),c)
print(add.numpy())
```

```python
X = tf.constant([[3,4,5],[1,2,3]])
Y = tf.constant([[4,5,6],[2,3,34]])
add = tf.add(X,Y)
print(add.numpy())
mul = tf.multiply(X,Y)
print(mul.numpy())
```

```python
x = tf.constant([[1.,1.],[2.,2.]])
a = tf.reduce_mean(x)
b = tf.reduce_mean(x,0) #axis 0
c = tf.reduce_mean(x,1) #axis 1
d = tf.reduce_mean(x, keepdims = True, axis=0)
print(a.numpy())
print(b.numpy())
print(c.numpy())
print(d.numpy())
```

```python
a = tf.constant([[10.0,20.0],[30.0,40.0]])
b = tf.constant([[20.0,30.0],[40.0,50.0]])
c = tf.constant([[30.0,40.0],[50.0,60.0]])
mysum = tf.reduce_sum([a,b,c])
print(mysum.numpy())
```

# Task 2: Perform linear algebra operations using Tensorflow


```python
A = tf.constant([[2,7],[3,4]])
At = tf.transpose(A)
print(A.numpy())
print(A.numpy())
```

# Task 3: Perform derivative and higher order derivative using gradient tape

f(n)= x^3





```python
#Constant
x= tf.constant(4.0)
with tf.GradientTape() as tape:
  tape.watch(x)
  y=x**3  ##f(x)
grad= tape.gradient(y,x)  #dy/dx
print(grad.numpy())
```

```python
#Variable
x= tf.Variable(4.0)
# x= tf.constant(4.0, trainable= True)    #ALSO WRITTEN AS
with tf.GradientTape() as tape:
  y=x**3  ##f(x)
grad= tape.gradient(y,x)  #dy/dx
print(grad.numpy())
```

```python
#Variable double derivative
x= tf.Variable(4.0)
with tf.GradientTape() as tape1:
  with tf.GradientTape() as tape2:
    y=x**3  ##f(x)
  grad_dy_dx= tape2.gradient(y,x)  #dy/dx #ORDER 1
grad_d2y_dx2= tape1.gradient(grad_dy_dx,x)  #d2y/dx2 #ORDER 2
print(grad_dy_dx.numpy())
print(grad_d2y_dx2.numpy())
```

# Task 4: Compute WX+b  where W,X,, and b  are drawn from a random normal distribution. W is of shape (4, 3), X is (3,1) and b is (4,1).

```python
W = tf.constant(np.random.randn(4,3))
X = tf.constant(np.random.randn(3,1))
b = tf.constant(np.random.randn(4,1))
output = tf.add(tf.matmul(W,X),b)
print(output)
```
